```python
def create_system_prompt(transcript, summary):
    """
    Generates a system prompt for evaluating the quality of an AI-generated summary of a doctor-patient conversation.
    """

    system_prompt = f"""
    You are a medical expert. Your task is to evaluate the accuracy and quality of an AI-generated summary of a doctor-patient conversation. Please provide an impartial assessment based on the following criteria, rating each from 1 to 5 (1 = lowest, 5 = highest). If a specific criterion is not addressed in the transcript, assign a score of 0.

    Carefully assess the summary using these criteria:

    Transcript:
    {transcript}

    Summary:
    {summary}

    Evaluation Criteria:
    1. Completeness of positive clinical signs, symptoms, or relevant details: Does the summary capture all important positive findings from the transcript?
    2. Completeness of negative clinical signs, symptoms, or relevant details: Does the summary capture all important negative findings from the transcript?
    3. Accuracy of diagnosis: Does the summary correctly reflect the diagnosis mentioned in the transcript?
    4. Accuracy of treatment plan: Does the summary correctly convey the treatment plan discussed in the transcript?
    5. Factual accuracy: Does the summary avoid any false, incorrect, or fabricated information?
    6. Clarity and structure: Is the summary concise, clear, and well-structured?

    Please return the evaluation scores in the following JSON format:
    {{
        "Criterion 1": score,
        "Criterion 2": score,
        "Criterion 3": score,
        "Criterion 4": score,
        "Criterion 5": score,
        "Criterion 6": score
    }}
    """

    return system_prompt
```

```python
# Generate prompts for each row
summary_data['system_prompt'] = summary_data.apply(lambda row: create_system_prompt(row['transcript'], row['summary']), axis=1)
summary_data
```

```python
prompts=summary_data['system_prompt'].tolist()
```

```python
len(prompts)
```

```python
import pandas as pd
import openai

# Set the API key and model name
OPENAI_API_KEY = 'Your API key'
openai.api_key = OPENAI_API_KEY  # Replace with your actual API key
MODEL = "o1-preview"  # Or "gpt-4-turbo" depending on availability

# Initialize an empty list to store the ratings
ratings = []


for prompt in prompts:

    try:
        # Create a completion request for each document
        completion = openai.chat.completions.create(
            model=MODEL,
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        # Extract and store the ratings in the list
        rating = completion.choices[0].message.content
        ratings.append(rating)

    except Exception as e:
        print(f"Error generating completion: {e}")
```

```python
summary_data['outputs'] = ratings
```

```python
import regex as re
def extract_scores(output_with_json):
    # Regular expression pattern to match numbers following "Criterion X": pattern
    pattern = r'"Criterion \d+": (\d+)'

    # Find all matches of the pattern (this will capture the numeric scores)
    scores = re.findall(pattern, output_with_json)

    # Convert the captured scores from strings to integers
    scores = [int(score) for score in scores]

    return scores
```

```python
# Apply the function to the 'model_response' column

summary_data['ratings_list'] = summary_data['outputs'].apply(extract_scores)
summary_data
```

# Calculate BERT Score

```python
#fill up nan rows with "Missing" to avoid errors
summary_data['summary'] = summary_data[['summary']].fillna("Missing")
summary_data
```

```python
from bert_score import score as bert_score
from rouge import Rouge


def compute_score(valid_data):

    scores = bert_score(
        valid_data["summary"].tolist(),
        valid_data["doctor_summary"].tolist(),
        lang="en",
        verbose=True,
    )
    p, r, f1 = scores[0], scores[1], scores[2]
    BERTScore_Precision, BERTScore_Recall, BERTScore_F1 = (
        p.numpy().mean(),
        r.numpy().mean(),
        f1.numpy().mean(),
    )
    valid_data["BERTScore_Precision"] = p.numpy()
    valid_data["BERTScore_Recall"] = r.numpy()
    valid_data["BERTScore_F1"] = f1.numpy()

    print("BERTScore Precision (mean):", BERTScore_Precision)
    print("BERTScore Recall (mean):", BERTScore_Recall)
    print("BERTScore F1 (mean):", BERTScore_F1)
    
```

```python
compute_score(summary_data)
```
